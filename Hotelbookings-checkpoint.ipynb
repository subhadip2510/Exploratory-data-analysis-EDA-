# Data Analysis and Visualisation

Cancelation percentage

```python
cancelled_per = df['is_canceled'].value_counts(normalize=True)
print(cancelled_per)

plt.figure(figsize = (5, 4))
plt.title('Reservation status count')
plt.bar(['Not canceled','Canceled'], df['is_canceled'].value_counts(),edgecolor = 'k', width = 0.7)
plt.show()
```

Most cancellation ratio:

```python
plt.figure(figsize = (8, 4))
ax1 = sns.countplot(x = 'hotel',hue = 'is_canceled', data = df, palette = 'Blues')
legend_labels,_=ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor = (1,1))
plt.title('Reservation status in different hotels', size = 20)
plt.xlabel('hotel')
plt.ylabel('number of reservations')
plt.legend(['not cancelled', 'canceled'])
plt.show()
```

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts()
```

```python
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts()
```

Price effect on City Hotel and Resort Hotel

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

plt.plot(resort_hotel.index, resort_hotel['adr'], label='Resort Hotel', color='blue', marker='o')
plt.plot(city_hotel.index, city_hotel['adr'], label='City Hotel', color='green', marker='s')

plt.title('Average Daily Rate Comparison between Resort and City Hotels', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Average Daily Rate (ADR)', fontsize=12)

plt.legend(fontsize=12)
plt.grid(True)

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Reservation Status Per Month

```python
df['month']= df['reservation_status_date'].dt.month
plt.figure(figsize = (16, 8))
ax1 = sns.countplot(x = 'month', hue = 'is_canceled', data = df)
plt.title('Reservation status per month', fontsize=16)
plt.xlabel('Month')
plt.ylabel('No. of reservations')
plt.legend(['not canceled', 'canceled'])
plt.show()
```

```python


plt.figure(figsize=(15, 8))


plt.title('Average Daily Rate per Month for Canceled Bookings', fontsize=20)


sns.barplot(x='month', y='adr', data=df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index())


plt.xlabel('Month', fontsize=14)
plt.ylabel('Average Daily Rate (ADR)', fontsize=14)
plt.legend(labels=['Canceled ADR'], fontsize=12)  
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
cancelled_data = df[df['is_canceled'] ==1 ]
top_10_country = cancelled_data['country'].value_counts()[:10]
plt.figure(figsize = (8,8))
plt.title('Top 10 countries with reservation canceled')
plt.pie(top_10_country, autopct = '%.2f', labels = top_10_country.index)
plt.show()
```
